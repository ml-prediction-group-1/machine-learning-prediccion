```python
# Exploración del Dataset: Top Pakistani Universities (Kaggle + Croissant)

import pandas as pd
import numpy as np
import mlcroissant as mlc
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
croissant_dataset = mlc.Dataset(
    'https://www.kaggle.com/datasets/ayeshaseherr/top-pakistani-universities/croissant/download'
)

record_sets = croissant_dataset.metadata.record_sets
record_sets
```

```python
df = pd.DataFrame(croissant_dataset.records(record_set=record_sets[0].uuid))
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.describe(include='all')
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Ranking"], kde=True)
plt.title("Distribución del Ranking")
plt.show()
```

```python
df.hist(figsize=(16, 12), bins=20)
plt.suptitle("Distribución de variables numéricas", fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Matriz de correlación")
plt.show()
```

```python
cols_categoricas = [col for col in df.columns if df[col].dtype == "object"]

for col in cols_categoricas:
    plt.figure(figsize=(10,5))
    sns.boxplot(x=df[col], y=df["Ranking"])
    plt.title(f"Ranking según {col}")
    plt.xticks(rotation=45)
    plt.show()
```

```python
print("Duplicados antes:", df.duplicated().sum())
df = df.drop_duplicates()
print("Duplicados después:", df.duplicated().sum())
```

```python
# Rellenar nulos (o quitarlos)
df_clean = df.dropna()

df_clean.info()
```

```python
df_clean.to_csv("../data/universities.csv", index=False)
print("Dataset limpio guardado en ../data/universities.csv")
```
